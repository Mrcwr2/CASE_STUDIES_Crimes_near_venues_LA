```python
import pandas as pd
import numpy as np
import csv
```

```python
crime_file = pd.read_csv("C:\\Users\\Charles.Robinson\\Desktop\\2022 Portfolio Project\\Project 2\\Crime Data\\Crime_Data_from_2020_to_Present.csv")
```

```python
crime_file = crime_file.rename(columns={"LAT":"Y", "LON":"X"})
crime_file.shape
```

```python
# import venue file
venue_file = pd.read_csv("C:\\Users\\Charles.Robinson\\Desktop\\2022 Portfolio Project\\Project 2\\Venues\\Major_Sport_Venues.csv")
```

```python
#filter city-los angeles in venue and sort by Y
venue_file_la = venue_file[venue_file['CITY'].str.contains ("Los Angeles")]
venue_file_la = venue_file_la.sort_values('Y')
print(venue_file_la.head(10))
print(venue_file_la)
```

```python
# for each
for x in range(len(venue_file_la)):
    # declare an empty dataframe called DF
    DF = pd.DataFrame()
    # take 1 row of the venue file
    one_venue = venue_file_la.iloc[[x]]
    # get the range of +/- 20 miles from the x, y (using 69 miles equalling 1 degree of latitude and 54.6 miles equals 1 degree of longitude https://www.usgs.gov/faqs/how-much-distance-does-degree-minute-and-second-cover-your-maps#:~:text=One%2Ddegree%20of%20longitude%20equals,one%20second%20equals%2080%20feet.)
    one_venue["max_x"] = one_venue["X"] + 20/54.6
    one_venue["min_x"] = one_venue["X"] - 20/54.6
    one_venue["max_y"] = one_venue["Y"] + 20/69
    one_venue["min_y"] = one_venue["Y"] - 20/69
    
    # get the range of +/- 10 miles from the x, y (using 69 miles equalling 1 degree of latitude and 54.6 miles equals 1 degree of longitude https://www.usgs.gov/faqs/how-much-distance-does-degree-minute-and-second-cover-your-maps#:~:text=One%2Ddegree%20of%20longitude%20equals,one%20second%20equals%2080%20feet.)
    one_venue["max_x_10"] = one_venue["X"] + 10/54.6
    one_venue["min_x_10"] = one_venue["X"] - 10/54.6
    one_venue["max_y_10"] = one_venue["Y"] + 10/69
    one_venue["min_y_10"] = one_venue["Y"] - 10/69
    
    # Filter to within a distance of the stadium
    crime_filtered = crime_file[crime_file["X"] > one_venue["min_x"].iloc[0]]
    crime_filtered = crime_filtered[crime_filtered["X"] < one_venue["max_x"].iloc[0]]
    crime_filtered = crime_filtered[crime_filtered["Y"] > one_venue["min_y"].iloc[0]]
    crime_filtered = crime_filtered[crime_filtered["Y"] < one_venue["max_y"].iloc[0]]
    
    #crime_filtered
    #crime_filtered.shape
    
    # Combine venue with crime data
    one_venue_repeated = pd.DataFrame(np.repeat(one_venue.values, len(crime_filtered), axis=0))
    one_venue_repeated.columns = one_venue.columns
    
    # reset index for joining
    crime_filtered = crime_filtered.reset_index()
    one_venue_repeated = one_venue_repeated.reset_index()
    
    venue_crime_data_combined = pd.concat([crime_filtered, one_venue_repeated], axis=1)    
    
    # join
    DF = pd.concat([DF, venue_crime_data_combined])
    venue_name = one_venue['NAME'].iloc[0]
    DF.to_csv(f'C:\\Users\\Charles.Robinson\\Desktop\\2022 Portfolio Project\\Project 2\\Final DF\\Filtered Crime Venue DF {venue_name}.csv', index = False)
    

print(DF.head(10))
```
